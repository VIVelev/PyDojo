## Load modules

```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

from dojo.dimred import TSNE
from sklearn.manifold import TSNE as SkTSNE
```

## Prepare the Data

```python
X1 = np.random.randn(10, 3)
X1 += 3
X1 *= 1

X2 = np.random.rand(10, 3)
X2 += -1
X2 *= 3

X = np.concatenate((X1, X2))
color = ['red'] * 10 + ['blue'] * 10

ax = plt.axes(projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], color=color)
plt.show()
```

## t-SNE

```python
tsne = TSNE(n_components=2, perplexity=10, n_iter=2000, verbose=True)
Y = tsne.fit_transform(X)
```

```python
plt.scatter(Y[:, 0], Y[:, 1])
plt.show()
```

## *Scikit-learn* t-SNE

```python
sk_tsne = SkTSNE(n_components=2, perplexity=5, n_iter=2000, verbose=True)
Y = sk_tsne.fit_transform(X)
```

```python
plt.scatter(Y[:, 0], Y[:, 1])
plt.show()
```
