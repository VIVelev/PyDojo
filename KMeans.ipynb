```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs
```

```python
from dojo.cluster import KMeans
```

```python
K = 5
X, y = make_blobs(n_samples=500, n_features=2, centers=K, cluster_std=0.6)
```

```python
def plot(y):
    for cl, marker, color in zip(range(K), ['^', 'o', 'x', 'h', 'p'], ['k', 'blue', 'red', 'green', 'purple']):
        plt.scatter(X[y == cl, 0], X[y == cl, 1], c=color, marker=marker)
        plt.grid()
    
    plt.show()
```

```python
plot(y)
```

```python
kmeans = KMeans(n_clusters=K).fit(X)
```

```python
plot(kmeans.clusters)
```
